#### Exercise 09. Are the average prices of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about them.

```python
# Average prices of Valdemorillo and Galapagar
valdemorillo = ds[ds['level5'] == 'Valdemorillo']
galapagar = ds[ds['level5'] == 'Galapagar']
avg_valdemorillo = valdemorillo['price'].mean()
avg_galapagar = galapagar['price'].mean()
print(f"Average price in Valdemorillo: {avg_valdemorillo}")
print(f"Average price in Galapagar: {avg_galapagar}")
print("Conclusion: The average prices are different." if avg_valdemorillo != avg_galapagar else "Conclusion: The average prices are the same.")
```

#### Exercise 10. Are the average prices per square meter (price/m2) of "Valdemorillo" and "Galapagar" the same? (★★☆)

Print both average prices and then write a conclusion about it.

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
# Average price per square meter in Valdemorillo and Galapagar
ds['pps'] = ds['price'] / ds['surface']
valdemorillo = ds[ds['level5'] == 'Valdemorillo']
galapagar = ds[ds['level5'] == 'Galapagar']
avg_pps_valdemorillo = valdemorillo['pps'].mean()
avg_pps_galapagar = galapagar['pps'].mean()
print(f"Average PPS in Valdemorillo: {avg_pps_valdemorillo}")
print(f"Average PPS in Galapagar: {avg_pps_galapagar}")
print("Conclusion: The average PPS are different." if avg_pps_valdemorillo != avg_pps_galapagar else "Conclusion: The average PPS are the same.")
```

#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

Hint: You can help yourself by reviewing the *multihist* demo of Matplotlib.

```python
# Normalize prices and plot histograms
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
for pop in ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']:
    pop_data = south_belt[south_belt['level5'] == pop]
    prices = pop_data['price'].values.reshape(-1, 1)
    normalized = scaler.fit_transform(prices)
    plt.hist(normalized, bins=20, alpha=0.5, label=pop)
plt.legend()
plt.title('Normalized Price Histograms')
plt.show()
```

The histograms show the distribution of normalized prices. Most prices are clustered around the mean, with some outliers in higher ranges.

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

#### Exercise 20. Make the same plot for 4 different populations (level5 column) and rearrange them on the same graph. You must use the subset obtained in Exercise 14 (★★☆)

```python
# Scatter plots for each population
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
populations = ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']
for i, pop in enumerate(populations):
    ax = axes[i//2, i%2]
    pop_data = south_belt[south_belt['level5'] == pop]
    ax.scatter(pop_data['surface'], pop_data['price'])
    ax.set_title(pop)
    ax.set_xlabel('Surface (m²)')
    ax.set_ylabel('Price (USD)')
plt.tight_layout()
plt.show()
```

#### Exercise 21. Make a plot of the coordinates (latitude and longitude columns) of the south belt of Madrid by color of each population (you must use the subset obtained in Exercise 14) (★★★★)

Execute the following cell, and then start coding in the next one. You must implement a simple code that transforms the coordinates columns in a Python dictionary (add more information if needed) and then add it to the map

```python
from ipyleaflet import Map, basemaps

# Map centered on south west part of Madrid
map = Map(center = (40.35, -3.8), zoom = 11.5, min_zoom = 1, max_zoom = 20, 
    basemap=basemaps.OpenStreetMap.Mapnik)
map
```

```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:
from ipyleaflet import Marker
colors = {'Fuenlabrada': 'red', 'Leganés': 'blue', 'Getafe': 'green', 'Alcorcón': 'orange'}

# Create a dictionary with coordinates and additional info
coordinates_dict = {}
for pop in ['Fuenlabrada', 'Leganés', 'Getafe', 'Alcorcón']:
    pop_data = south_belt[south_belt['level5'] == pop].dropna(subset=['latitude', 'longitude'])
    coordinates_dict[pop] = pop_data[['latitude', 'longitude', 'address', 'price']].to_dict('records')

# Add markers to the map
for pop, coords in coordinates_dict.items():
    for item in coords:
        marker = Marker(location=(item['latitude'], item['longitude']), 
                        popup=f"{pop}: {item['address']} - {item['price']} USD")
        map.add_layer(marker)

map
```
